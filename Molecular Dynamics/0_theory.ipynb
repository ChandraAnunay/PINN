### 1. Molecular Dynamics

From a theoretical POV, one can know everything about a material using Schrodinger's equation. But computationally, the task becomes impossible. Thus, MD makes a simplifying approximation - 

" Treat the nuclei as classical particles moving under forces generated by an interatomic potential "

Key equation in MD :

$$ m_i\dfrac{d^2\vec{r}_i}{dt^2} = \vec{F}_i = -\vec{\nabla}_i U(\vec{r}_1,\vec{r}_2,...,\vec{r}_N) $$

Here '$i$' is the particle index in an $N$-particle system.

### 2. Born-Oppenheimer Approximation :

It says that since the mass of the nuclei is very large in comparision to the mass of electron, the nuclei almost don't change their positions that much as compared to the electrons. Thus, we can assume a static picture of nuclei in the atom.

This gives :

$$ H_e(\vec{R})\psi_e = E(\vec{R})\psi_e $$

Thus, instead of guessing the wavefunctions of millions of atoms, we just need to guess the multidimensional energy function : $ U(\vec{R}) = E(\vec{R}) $. This function is called the 'POTENTIAL ENERGY SURFACE (PES)'.

### 3. Methods of computing PES :

1. Using Quantum Mechanics (DFT) : Using Schrodinger equation to solve for the electronic wavefunction : ( Very accurate, but very slow )
2. Mathematical Modelling using Classical Potentials : eg. Lennard-Jones, Morse, EAM, Tersoff, etc. ( Less accurate, but fast )
3. Using Machine Learning : ( Fast and accurate )

### 4. MD pipeline :

A complete molecular dynamics pipeline looks like :

Forces obtained from -> Energy -> Energy computed using the above 3 methods -> $\vec{F} = -\vec{\nabla}U $-> Newton's equations -> Integrate in time (velocity verlet) -> Atomic Trajectories -> Material Properties

#### Exercise 1. Lennard-Jones Potential

$$ U(r) = 4\epsilon\left[ \left(\dfrac{\sigma}{r}\right)^{12} - \left(\dfrac{\sigma}{r}\right)^6 \right] $$

Differentiating w.r.t. $r$ we get the force as :

$$ F = 4\epsilon\left[ \dfrac{12}{r}\left(\dfrac{\sigma}{r}\right)^{12} - \dfrac{6}{r}\left(\dfrac{\sigma}{r}\right)^6 \right] $$
