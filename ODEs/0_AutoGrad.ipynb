```python
import torch
```

```python
a = torch.tensor([3.0],requires_grad=True)
b = torch.tensor([5.0],requires_grad=True)
c = a*b
d = a**2+b
print(f'a : {a}')
print(f'b : {b}')
print(f'c : {c}')
print(f'd : {d}')
```

1. Calling AutoGrad

```python
print(f"Before Calling _.backward()")
print(f"dcda = {a.grad}")
c.backward()
print(f"\nAfter Calling _.backward()")
print(f"dcda = {a.grad}")
```

2. Working with correct gradients

```python
# Similarly

print(f"dcdb = {b.grad}")
d.backward()
print(f"ddda = {a.grad}")
print(f"dddb = {b.grad}")
```

Notice that the gradients of $d$ are both wrong. This happens because it adds the previous gradients w.r.t. $a$ and $b$. To avoid this, we must go as below :

```python
a.grad = None
b.grad = None
d = a**2 + b
d.backward()
print(f"ddda = {a.grad}")
print(f"dddb = {b.grad}")
```
